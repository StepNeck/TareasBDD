## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Usuarios"

```python
%%sql

CREATE TABLE Usuarios (
    ID_usuario INTEGER GENERATED ALWAYS AS IDENTITY ,
    nombre VARCHAR(45) NOT NULL,
    apellido VARCHAR(45),
    correo VARCHAR(45) NOT NULL,
    contraseña VARCHAR(50) NOT NULL,
    fecha_registro TIMESTAMP NOT NULL,
    PRIMARY KEY (ID_usuario)
);
```

### Tabla "Creadores"

```python
%%sql

CREATE TABLE Creadores(
    ID_creador INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    apellido VARCHAR(45) NOT NULL,
    popularidad INTEGER,
    estilo VARCHAR(45),
    PRIMARY KEY (ID_creador)
);
```

### Tabla "Tipos_NTF"

```python
%%sql

CREATE TABLE Tipos_NFT(
    ID_tipo INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(20) NOT NULL,
    PRIMARY KEY (ID_tipo)
);
```

### Tabla "NFT"

```python
%%sql
CREATE TABLE NFT(
    ID_nft INTEGER GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    ID_creador INTEGER NOT NULL,
    ID_tipo INTEGER,
    PRIMARY KEY (ID_nft),
    FOREIGN KEY (ID_creador) REFERENCES Creadores (ID_creador),
    FOREIGN KEY (ID_tipo) REFERENCES Tipos_nft (ID_tipo)
);
```

### Tabla "Precio NFT"

```python
%%sql

CREATE TABLE Precio_NFT(
    ID_nft INTEGER,
    fecha TIMESTAMP NOT NULL,
    precio INT NOT NULL,
    PRIMARY KEY (ID_nft, fecha),
    FOREIGN KEY (ID_nft) REFERENCES NFT (ID_nft)
);
```

### Tabla "Usuario tiene NFT"

```python
%%sql

CREATE TABLE Usuario_tiene_NFT(
    ID_usuario INTEGER,
    ID_nft INTEGER,
    PRIMARY KEY (ID_usuario, ID_nft),
    FOREIGN KEY (ID_usuario) REFERENCES Usuarios (ID_usuario),
    FOREIGN KEY (ID_nft) REFERENCES NFT (ID_nft)
);
```

### Consulta 2
**Cantidad de precios que ha tenido históricamente cada NTF.**

Ejemplo:


| NFT     | Cantidad de precios |
|--------------|----------|
| Minecraft Sphere |    3     |
| The Binary Girl     |    7     |
| ...          |   ...    |

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql

SELECT 
    Tabla.NFT, Tabla.Cantidad_de_Precios
FROM(
    SELECT
        NFT.nombre aS NFT, final.Cantidad as Cantidad_de_Precios
    FROM(
        SELECT
            Tabla1.ides as identificador,
            count(*) as Cantidad
        FROM(
            SELECT
                ID_nft as ides
            FROM
                Precio_NFT
        )as Tabla1

        GROUP BY
            Tabla1.ides
    )as final INNER JOIN NFT

    ON
        final.identificador=NFT.ID_nft
    
)as Tabla;
```

### Consulta 3
**El nombre y tipo (nombre del tipo) del NFT históricamente más caro.**
###### Si hay más de uno, ponga solamente el primero
Ejemplo:

| NFT    | Tipo | 
|----------------|--------|
| Minecraft Sphere| Arte 3D |

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
SELECT
    Tabla.NFT,Tabla.Tipo
FROM(
    
SELECT
    final.nombrecito as NFT,Tipos_NFT.nombre as Tipo
    

FROM(
    SELECT
        NFT.nombre as nombrecito,
        NFT.ID_tipo as identificador

    FROM(

        SELECT
            Precio_NFT.ID_nft as laid,
            maximo.preciomax as maxi
    

        FROM (
            SELECT
                max(precio) as preciomax
            FROM
                Precio_NFT
            ) as maximo INNER JOIN Precio_NFT

        ON
            maximo.preciomax=Precio_NFT.precio
    
    ) as penultimo INNER JOIN NFT

    ON 
        penultimo.laid=NFT.ID_nft
    
)as final INNER JOIN Tipos_NFT

ON
    final.identificador=Tipos_NFT.ID_tipo
    
)as Tabla;
```

### Consulta 4
**Nombre y apellido del usuario con más número de NFTs.**

Ejemplo:

| Nombre     | Apellido |
|----------------|--------|
| Levi | Ackerman |

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql

SELECT 
    Usuarios.nombre, Usuarios.apellido
FROM(
    SELECT 
        Tabla3.usua as usuari, Tabla3.canti as cantidad
    FROM(
        SELECT 
            Tabla2.idusua as usua, Tabla2.veces as canti
        FROM(
            SELECT
                Tabla1.idusuario as idusua, count(*) as veces
            FROM(
                SELECT
                    ID_Usuario as idusuario ,ID_nft 
                FROM 
                    Usuario_tiene_NFT
            )as Tabla1

            GROUP BY
                Tabla1.idusuario
        )as Tabla2

        ORDER BY
            Tabla2.veces DESC
    )as Tabla3
    
    LIMIT 1
)as final INNER JOIN Usuarios

ON
    final.usuari=Usuarios.ID_usuario
```

### Consulta 5
**El NFT que más veces cambió de precio historicamente.**

Ejemplo:

|  ID  | NFT | 
|--------|-------|
| 11 |Minecraft Sphere | 

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql

SELECT 
    Tabla.id_nft, Tabla.NFT
FROM(
    SELECT
        NFT.nombre as NFT, final.Cantidad as Cantidad_de_Precios, NFT.ID_nft as id_nft
    FROM(
        SELECT
            Tabla1.ides as identificador,
            count(*) as Cantidad
        FROM(
            SELECT
                ID_nft as ides
            FROM
                Precio_NFT
        )as Tabla1

        GROUP BY
            Tabla1.ides
    )as final INNER JOIN NFT

    ON
        final.identificador=NFT.ID_nft
)as Tabla

ORDER BY
    Tabla.cantidad_de_precios DESC
LIMIT 1;
```

### Consulta 7
**Top 4 NFTs más caros actualmente de un id_tipo X .**

|Minecraft Sphere | 
|The Binary Girl |
|Koalas in Love |

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
SELECT
    nombre
FROM (
    SELECT
        tabla1.id_nft, tabla1.fecha_nft, tabla1.precio_nft
    FROM (
        SELECT DISTINCT ON (ID_nft)
            ID_nft as id_nft, fecha as fecha_nft, precio as precio_nft
        FROM
            Precio_NFT
        ORDER BY
            ID_nft, fecha DESC
    )as tabla1
    ORDER BY 
        tabla1.precio_nft DESC
)as tabla INNER JOIN NFT ON tabla.id_nft = NFT.ID_nft
LIMIT 4;
```
